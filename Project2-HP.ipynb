```python
# One hot encoding
hp=pd.get_dummies(data=hp,columns=['ocean_proximity'],drop_first=True)
```

All null values are cleared and all the object data-types are converted into numerical

```python
hp
```

```python
hp.rename(columns={'ocean_proximity_NEAR BAY':'ocean_proximity_NEAR_BAY','ocean_proximity_NEAR OCEAN':'ocean_proximity_NEAR_OCEAN','ocean_proximity_<1H OCEAN':'ocean_proximity_lessthan_1H_OCEAN'},inplace=True)
```

```python
hp_x=hp.drop(labels='median_house_value',axis=1)
hp_x
```

```python
hp_y=hp["median_house_value"]
hp_y
```

```python
# Standardizing Data
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
# Train Test Spit
from sklearn.model_selection import train_test_split
hp_x_train , hp_x_test  , hp_y_train , hp_y_test = train_test_split(hp_x , hp_y , test_size = .3)
```

```python
hp_x_train=scalar.fit_transform(hp_x_train)
```

```python
hp_x_test =scalar.transform(hp_x_test)
```

```python
# Model Building using Linear regression
from sklearn.linear_model import LinearRegression
le=LinearRegression()
```

```python
le.fit(hp_x_train,hp_y_train)
```

```python
le.score(hp_x_train,hp_y_train)
```

```python
le.score(hp_x_test,hp_y_test)
```

```python
hp_y_pred=le.predict(hp_x_test)
```

```python
hp_y_pred
```

```python
hp_y_test
```

```python
df={"actual":hp_y_test,"predicted":hp_y_pred}
compare=pd.DataFrame(df)
```

```python
compare
```

```python
compare.predicted =[round(i) for i in compare.predicted]
```

```python
compare
```

```python
# Checking wither the error terms are normally distributed with a mean of zero
compare["errors"]=compare.actual-compare.predicted
compare
```

```python
plt.figure(figsize=(8, 5))
plt.hist(compare["errors"],bins=100)
plt.title("Error Distribution");
```

```python
plt.figure(figsize=(8, 4), dpi=120, edgecolor='b')
x=[i for i in range(6192)]
y=[0 for i in x]
plt.scatter( x, compare.errors, label = 'Errors')
plt.plot(x,y,color='r',label='Mean')
plt.title("Error Variation about the Mean")
plt.legend(loc='lower left');
```

The errors are normally distributed

```python
from sklearn import metrics
mse=metrics.mean_squared_error(hp_y_test,hp_y_pred)
rmse=np.sqrt(mse)
rmse
```

```python
# Model building using Ridge and Lasso regression
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(hp_x_train , hp_y_train)
ridge.score(hp_x_test , hp_y_test)
hp_y_pred=ridge.predict(hp_x_test)
mse=metrics.mean_squared_error(hp_y_test,hp_y_pred)
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(hp_x_train , hp_y_train)
hp_y_pred=lasso.predict(hp_x_test)
mse=metrics.mean_squared_error(hp_y_test,hp_y_pred)
rmse=np.sqrt(mse)
rmse
```

```python
# Model building using RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(hp_x_train, hp_y_train)
hp_y_pred=rf.predict(hp_x_test)
mse=metrics.mean_squared_error(hp_y_test,hp_y_pred)
rmse=np.sqrt(mse)
rmse
```

```python
# Model building using neural network
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Activation, Dense 
```

```python
model = Sequential()
model.add(Dense(12, input_dim=12, kernel_initializer='normal', activation='relu'))
model.add(Dense(12, kernel_initializer='normal', activation='relu'))
model.add(Dense(24, kernel_initializer='normal', activation='relu'))
model.add(Dense(48, kernel_initializer='normal', activation='relu'))
model.add(Dense(48, kernel_initializer='normal', activation='relu'))
model.add(Dense(48, kernel_initializer='normal', activation='relu'))
model.add(Dense(48, kernel_initializer='normal', activation='relu'))
model.add(Dense(48, kernel_initializer='normal', activation='relu'))
model.add(Dense(20, kernel_initializer='normal', activation='relu'))
model.add(Dense(20, kernel_initializer='normal', activation='relu'))
model.add(Dense(20, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='relu'))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(hp_x_train , hp_y_train, batch_size=10, epochs=20, validation_split = 0.3)
```

```python
hp_y_pred=model.predict(hp_x_test, batch_size=20)
mse=metrics.mean_squared_error(hp_y_test,hp_y_pred)
rmse=np.sqrt(mse)
rmse
```
